#### Tahap: Standardisasi Fitur Numerik

Pada tahap ini, dilakukan standardisasi pada fitur numerik menggunakan `StandardScaler` untuk mengubah skala data sehingga memiliki mean 0 dan standar deviasi 1. Langkah ini penting agar fitur numerik berada dalam rentang yang sebanding, membantu algoritma machine learning bekerja lebih efektif.


```python
# Select numerical features for scaling
numerical_features = data.select_dtypes(include=np.number).columns.tolist()

# Initialize the scaler (choose either StandardScaler or MinMaxScaler)
# StandardScaler for standardization (mean=0, std=1)
scaler_standard = StandardScaler()

# Apply standardization to the selected numerical features
data[numerical_features] = scaler_standard.fit_transform(data[numerical_features])

print("\nData after Standardization:")
print(data.head())
```

> Hasil standardisasi memperlihatkan nilai fitur numerik yang telah ditransformasi ke skala standar, dengan distribusi yang telah diseimbangkan. Fitur kategorikal tetap tidak berubah. Data siap untuk tahap pemodelan berikutnya.


#### Tahap: Penanganan Outlier dengan Metode IQR

Pada tahap ini, dilakukan deteksi dan penanganan outlier pada fitur numerik menggunakan metode Interquartile Range (IQR). Data yang berada di luar batas bawah dan atas IQR dianggap outlier dan akan dihapus untuk meningkatkan kualitas data.


```python
def handle_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_outlier_column = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)].copy()
    return df_outlier_column

data_cleaned = data.copy()
for col in data.select_dtypes(include=np.number).columns:
    data_cleaned = handle_outliers_iqr(data_cleaned, col)

print(f"Original shape: {data.shape}")
print(f"Shape after outlier removal: {data_cleaned.shape}")
```

> Hasil menunjukkan pengurangan jumlah data dari 4269 menjadi 4173 baris setelah penghapusan outlier, yang menandakan sejumlah data ekstrem telah berhasil dibersihkan sehingga dataset lebih representatif untuk analisis dan pemodelan selanjutnya.


#### Tahap: Encoding Fitur Kategorikal dengan LabelEncoder

Pada tahap ini, fitur kategorikal diubah menjadi representasi numerik menggunakan metode Label Encoding. Transformasi ini diperlukan agar algoritma machine learning dapat memproses data kategorikal secara efektif.


```python
# Encoding categorical features
categorical_features = data_cleaned.select_dtypes(include='object').columns.tolist()

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Apply Label Encoding to each categorical feature
for col in categorical_features:
    data_cleaned[col] = label_encoder.fit_transform(data_cleaned[col])

print("\nData after Label Encoding:")
print(data_cleaned.head())
```

> Hasil output menunjukkan bahwa fitur kategorikal seperti `education`, `self_employed`, dan `loan_status` telah berhasil diubah menjadi nilai numerik. Data siap untuk digunakan dalam proses pelatihan model selanjutnya.
